```python
# ---------------------------------------------------------------------
# Principal
# ---------------------------------------------------------------------

import random
import math

# Inicializamos la lista sin numeros repetidos
#random.seed(42) # Una semilla para que tus números sean iguales a los míos
limite_max = 1000
inc = 7
lista_1d = [random.randrange(1, limite_max) for i in range(inc)]
lista_ordenada = sorted(lista_1d)
#print(lista_1d)
print()


# Lista de puntos 2D de ejemplo /  Comprensión de Listas (List Comprehension).
# Tupla (X, Y): Python crea una estructura inmutable llamada Tupla. Esto representa un único punto fijo en tu plano cartesiano.
lista_2d = [(random.randrange(1, limite_max), random.randrange(1, limite_max)) for i in range(inc)] # el for indica : Cuántos puntos queremos
#print(lista_2d)
print()


lista_3d = [(random.randrange(1, limite_max), random.randrange(1, limite_max), random.randrange(1, limite_max)) for i in range(inc)]
print(lista_3d)
print()






def distancia_1d(a, b):
    # La distancia en una línea recta: el valor absoluto de la resta
    return abs(a - b)
```

```python
# ---------------------------------------------------------------------
# CASO 1: 1D - Método (Fuerza Bruta)
# Encontrar los dos números más cercanos de una lista.
# ---------------------------------------------------------------------
def fuerza_bruta_1d(lista):
    # Cualquier distancia real será menor que infinito,
    # permite guardar la primera pareja
    # de inmediato.
    distancia_minima = float('inf') # Empezamos con una distancia infinita
    pareja_mas_cercana = None
    n = len(lista) # Longitud de la lista
    comparaciones = 0 # Cuenta número de Comparaciones

    # Bucle exterior: Elige el primer número (i)
    for i in range(n):
        # Bucle interior: compara con los números siguientes (j)
        for j in range(i + 1, n): # evita comparar un número consigo mismo
            comparaciones += 1
            d = distancia_1d(lista[i], lista[j])

            # Si encontramos una distancia más pequeña, la guardamos
            if d < distancia_minima:
                distancia_minima = d
                pareja_mas_cercana = (lista[i], lista[j])

    return pareja_mas_cercana, distancia_minima, comparaciones
```

```python
# ---------------------------------------------------------------------
# CASO 2: 1D - Método (Fuerza Bruta) OPTIMIZADO
# Encontrar los dos números más cercanos de una lista.
# ---------------------------------------------------------------------

def optimizado_ordenado_1d(lista):
    # Paso 1: Ordenamos la lista para agrupar los números más cercanos.
    # Usamos sorted() para no modificar la lista original
    #lista_ordenada = sorted(lista)
   # Cualquier distancia real será menor que infinito,
    # permite guardar la primera pareja
    # de inmediato.
    distancia_minima = float('inf')
    pareja_mas_cercana = None
    n = len(lista_ordenada)
    comparaciones = 0

    # Paso 2: Un solo bucle lineal.
    # Comparamos cada número (i) únicamente con su vecino de al lado (i + 1).
    # Recorremos hasta (n - 1) para que el último elemento no busque un vecino inexistente.
    for i in range(n - 1):
        comparaciones += 1
        d = distancia_1d(lista_ordenada[i], lista_ordenada[i + 1])

        # Si encontramos una distancia menor, la registramos
        if d < distancia_minima:
            distancia_minima = d
            # Guardamos la pareja correspondiente de la lista ordenada
            pareja_mas_cercana = (lista_ordenada[i], lista_ordenada[i + 1])

    return pareja_mas_cercana, distancia_minima, comparaciones
```

```python
# ---------------------------------------------------------------------
# CASO 3: 1D - Método Divide y Venceras (Recursividad)
# Encontrar los dos números más cercanos de una lista.
# ---------------------------------------------------------------------

def divide_y_venceras_1d_nivel(lista_ordenada, nivel=1):
                              #La expresión nivel = 1 en la cabecera de la función es un argumento por defecto (o valor de respaldo).
                              #Significa: "Si el programador NO  envía un valor para el parámetro nivel, entonces usa el número 1
                              #Pero si el programador SÍ envía un valor, Python ignora por completo el = 1 y utiliza el número que le acabas de mandar.



    # 📢 AVISO DE ENTRADA: Nos dice qué nivel se abre y con qué datos trabajar
    print(f" ENTRANDO a [Layer Nivel {nivel}] | Procesando sublista: {lista_ordenada}")
    #print("Press Enter to continue...")
    #input() # This will pause the program until Enter is pressed



    n = len(lista_ordenada)
    #print('Lista Ordenada --->',lista_ordenada )
    # Caso base: Si hay 2 o 3 elementos, los resolvemos directamente
    if n <= 3: # Un algoritmo recursivo necesita saber cuándo detenerse para no quedarse en un bucle infinito.
        # Buscamos el par más cercano por fuerza bruta Pero una lista pequeña
        min_d = float('inf') #Compara infinitos
        par = None
        for i in range(n):
            for j in range(i + 1, n):
                #d = abs(lista_ordenada[i] - lista_ordenada[j])
                d = distancia_1d(lista_ordenada[i], lista_ordenada[j])
                if d < min_d:
                    min_d = d
                    par = (lista_ordenada[i], lista_ordenada[j])


        # 📢 AVISO DE SALIDA DE CASO BASE: Nos dice qué encontró el hijo antes de cerrarse
        print(f"{nivel}     [Caso Base Activo] -> Par local: {par} | Distancia: {min_d}")
        print(f"{nivel} SALIENDO de [Layer Nivel {nivel}]")
        return par, min_d

    # 1. DIVIDIR: Encontrar el índice del medio
    medio = n // 2  # divide la lista entre 2 partes Iz. y Der.
    izq = lista_ordenada[:medio]
    der = lista_ordenada[medio:]
    print(f"   [Acción] Cortando por la mitad -> Izq: {izq} | Der: {der}")

    # 2. VENCER: Llamadas recursivas para ambas mitades
    print(f"   [Llamada] Función del Nivel {nivel} llama a su Hijo Izquierdo...")
    par_izq, d_izq = divide_y_venceras_1d_nivel(izq, nivel + 1) # Aumentamos nivel

    print(f"   [Llamada] Función del Nivel {nivel} llama a su Hijo Derecho...")
    par_der, d_der = divide_y_venceras_1d_nivel(der, nivel + 1) # Aumentamos nivel

    # Nos quedamos con el mínimo de las dos mitades
    if d_izq < d_der:
        delta = d_izq
        mejor_par = par_izq
    else:
        delta = d_der
        mejor_par = par_der

    # 3. COMBINAR: Comprobar la frontera (último de izq vs primero de der)
    punto_izq = izq[-1]
    punto_der = der[0]
    distancia_cruce = abs(punto_izq - punto_der)
    print(f"     [Evaluación Central] Cruzando frontera -> |{punto_izq} - {punto_der}| = {distancia_cruce}")
    if distancia_cruce < delta:
        delta = distancia_cruce
        mejor_par = (punto_izq, punto_der)
        print(f"     ¡El cruce ganó! Nuevo mínimo temporal: {delta}")
    return mejor_par, delta
```

```python
# ---------------------------------------------------------------------
# CASO 4: EL ESPACIO 2D - Puntos en el plano (x, y)
# ---------------------------------------------------------------------

# En dos dimensiones, los números se convierten en coordenadas espaciales:
# un punto se representa como una tupla $(x, y)$.
#Debemos aplicar el Teorema de Pitágoras para calcular la distancia en línea recta



def distancia_2d(p1, p2):
    # Distancia Euclidiana en dos dimensiones (Pitágoras)
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def fuerza_bruta_2d(puntos):
    min_d = float('inf')
    par = None
    n = len(puntos)
    for i in range(n):
        for j in range(i + 1, n):
            d = distancia_2d(puntos[i], puntos[j])
            if d < min_d:
                min_d = d
                par = (puntos[i], puntos[j])
    return par, min_d

def buscar_en_franja(franja, delta):
    # Ordenamos la franja vertical por la coordenada Y (de abajo hacia arriba)
    franja.sort(key=lambda punto: punto[1])
    min_d = delta
    mejor_par = None
    n = len(franja)

    # Escaneamos comparando con los vecinos de arriba
    for i in range(n):
        for j in range(i + 1, n):
            if (franja[j][1] - franja[i][1]) >= min_d:
                break
            d = distancia_2d(franja[i], franja[j])
            if d < min_d:
                min_d = d
                mejor_par = (franja[i], franja[j])
    return mejor_par, min_d

def par_mas_cercano_2d_recursivo(puntos_ordenados_x):
    n = len(puntos_ordenados_x)
    if n <= 3:
        return fuerza_bruta_2d(puntos_ordenados_x)

    # DIVIDIR
    medio = n // 2
    izq = puntos_ordenados_x[:medio]
    der = puntos_ordenados_x[medio:]

    # VENCER (Recursión)
    par_izq, d_izq = par_mas_cercano_2d_recursivo(izq)
    par_der, d_der = par_mas_cercano_2d_recursivo(der)
    # ejemplo Mitad Izquierda            │          Mitad Derecha
    #                                    │
    #               (135, 407)           │
    #                                    │
    #                     (290, 84)      │               (512, 280)
    #                                    │
    #                              Muro Central

    if d_izq < d_der:
        delta = d_izq
        mejor_par = par_izq
    else:
        delta = d_der
        mejor_par = par_der

    # COMBINAR (Frontera)
    punto_division = puntos_ordenados_x[medio]
    franja = [p for p in puntos_ordenados_x if abs(p[0] - punto_division[0]) < delta]

    par_franja, d_franja = buscar_en_franja(franja, delta)

    if par_franja is not None and d_franja < delta:
        return par_franja, d_franja

    return mejor_par, delta

def par_mas_cercano_2d(puntos):
    # Pre-ordenamiento estratégico en el eje X para habilitar Divide y Vencerás
    puntos_ordenados = sorted(puntos, key=lambda p: p[0])
    #Ejemplo
    #Antes de ordenar:  [(512, 280), (135, 407), (290, 84)]
    #Después de ordenar: [(135, 407), (290, 84), (512, 280)]
    #                    X=135       X=290       X=512

    #Encapsulación y usabilidad:
    # Devuelve el resultado final que esa función calcule.
    return par_mas_cercano_2d_recursivo(puntos_ordenados) #Delegación de Funciones o Función Envoltorio (Wrapper)
    # De lo contrario seria
    #lista_preparada = sorted(lista_2d, key=lambda p: p[0])
    #pareja, dist = par_mas_cercano_2d_recursivo(lista_preparada)




```

```python
# ---------------------------------------------------------------------
# CASO 5: Lista de puntos 3D de ejemplo (x, y, z)
# Puntos en el Plano
# ---------------------------------------------------------------------

# El Teorema de Pitágoras en 3D: distancia_3d
# Para calcular la distancia real en línea recta entre dos puntos en un espacio tridimensional
# P_1(x_1, y_1, z_1) y P_2(x_2, y_2, z_2), extendemos la distancia euclidiana añadiendo la diferencia del eje z:
# En Python, mapeando las variables, queda de la siguiente manera:
# p1[0] y p2[0] -> Eje X
# p1[1] y p2[1] -> Eje Y
# p1[2] y p2[2] -> Eje Z

def distancia_3d(p1, p2):
    print(p1)
    print(p2)
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

def buscar_en_caja_3d(caja, delta):
    # Ordenamos la caja central por el eje Y (altura)
    caja.sort(key=lambda punto: punto[1])
    min_d = delta
    mejor_par = None
    n = len(caja)

    for i in range(n):
        for j in range(i + 1, n):
            # 1. Optimización en Y: Si la diferencia de altura supera el récord, rompemos
            if (caja[j][1] - caja[i][1]) >= min_d:
                break

            # 2. Optimización en Z: Si la diferencia de profundidad supera el récord, ignoramos este punto
            if abs(caja[j][2] - caja[i][2]) >= min_d:
                continue

            d = distancia_3d(caja[i], caja[j])
            if d < min_d:
                min_d = d
                mejor_par = (caja[i], caja[j])
    return mejor_par, min_d



def par_mas_cercano_3d_recursivo(puntos_ordenados_x):
    n = len(puntos_ordenados_x)

    # Caso base
    if n <= 3:
        return fuerza_bruta_3d(puntos_ordenados_x)

    # 1. DIVIDIR: Plano divisor por la mitad del eje X
    medio = n // 2
    izq = puntos_ordenados_x[:medio]
    der = puntos_ordenados_x[medio:]

    # 2. VENCER: Recursión para cada mitad del espacio
    par_izq, d_izq = par_mas_cercano_3d_recursivo(izq)
    par_der, d_der = par_mas_cercano_3d_recursivo(der)

    if d_izq < d_der:
        delta = d_izq
        mejor_par = par_izq
    else:
        delta = d_der
        mejor_par = par_der

    # 3. COMBINAR: Caja tridimensional central de ancho 2 * delta
    punto_division = puntos_ordenados_x[medio]
    caja = [p for p in puntos_ordenados_x if abs(p[0] - punto_division[0]) < delta]

    par_caja, d_caja = buscar_en_caja_3d(caja, delta)

    if par_caja is not None and d_caja < delta:
        return par_caja, d_caja

    return mejor_par, delta

def par_mas_cercano_3d(puntos):
    # Pre-ordenación inicial obligatoria en el eje X para habilitar la división
    puntos_ordenados = sorted(puntos, key=lambda p: p[0])
    return par_mas_cercano_3d_recursivo(puntos_ordenados)
```
