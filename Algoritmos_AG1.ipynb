 # 🗼 Towers of Hanoi: Step-by-Step Execution Trace

Below is a detailed walkthrough of how a **3-disk stack** (Large, Medium, Small) transitions from its original source peg to its final destination peg using a recursive strategy.

---

### 🎬 Starting Position
* **Peg 1:** `[ Large, Medium, Small ]` *(The full initial stack)*
* **Peg 2:** `[ Empty ]`
* **Peg 3:** `[ Empty ]`

---

### 🕒 Phase 1: Clearing the Top Disks
* **Step 1:** Move **Small** from **Peg 1** to **Peg 3**
  * **Peg 1:** `[ Large, Medium ]`
  * **Peg 2:** `[ Empty ]`
  * **Peg 3:** `[ Small ]`

* **Step 2:** Move **Medium** from **Peg 1** to **Peg 2**
  * **Peg 1:** `[ Large ]`
  * **Peg 2:** `[ Medium ]`
  * **Peg 3:** `[ Small ]`

* **Step 3:** Move **Small** from **Peg 3** to **Peg 2** *(Stacks on top of Medium)*
  * **Peg 1:** `[ Large ]`
  * **Peg 2:** `[ Medium, Small ]`
  * **Peg 3:** `[ Empty ]` 🔓 *(Peg 3 is now perfectly clear!)*

---

### 👑 Phase 2: Moving the Core Base
* **Step 4: The Major Milestone (Base Case)**
  * **Action:** Move **Large** from **Peg 1** to **Peg 3**
  * **Peg 1:** `[ Empty ]`
  * **Peg 2:** `[ Medium, Small ]`
  * **Peg 3:** `[ Large ]` 🏠 *(The foundation disk has reached its permanent home!)*

---

### 🕒 Phase 3: Rebuilding the Stack
* **Step 5:** Move **Small** from **Peg 2** to **Peg 1**
  * **Peg 1:** `[ Small ]`
  * **Peg 2:** `[ Medium ]`
  * **Peg 3:** `[ Large ]`

* **Step 6:** Move **Medium** from **Peg 2** to **Peg 3** *(Stacks on top of Large)*
  * **Peg 1:** `[ Small ]`
  * **Peg 2:** `[ Empty ]`
  * **Peg 3:** `[ Large, Medium ]`

---

### 🏁 Final State
* **Step 7: The Final Move**
  * **Action:** Move **Small** from **Peg 1** to **Peg 3** *(Stacks on top of Medium)*
  * **Peg 1:** `[ Empty ]`
  * **Peg 2:** `[ Empty ]`
  * **Peg 3:** `[ Large, Medium, Small ]`  ✅ *(Puzzle complete!)*

## 🧮 Algorithmic Complexity & Mathematical Pattern

The Towers of Hanoi utilizes the **Divide and Conquer** paradigm through recursion. The mathematical relationship governing the total number of moves can be analyzed through two distinct lenses:

---

### 1. The Operational Logic (Recurrence Relation)
Every time a new disk is added to the tower, the system must repeat the work required for the previous stack size exactly **twice** (once to clear it out of the way, and once to stack it back on top), plus **one** single move for the new base disk.

$$\text{Total Work for New Size} = (\text{Work for Previous Size} \times 2) + 1$$

* **1 Disk:** $1$ move
* **2 Disks:** $(1 \times 2) + 1 = 3$ moves
* **3 Disks:** $(3 \times 2) + 1 = 7$ moves
* **4 Disks:** $(7 \times 2) + 1 = 15$ moves

---

### 2. The Shortcut Formula (Explicit Equation)
Because the work doubles sequentially with each addition, the growth directly mirrors powers of two ($2, 4, 8, 16...$), minus $1$. For any given number of disks ($N$), the exact total move count is calculated using the following mathematical expression:

$$\text{Total Moves} = 2^N - 1$$

* **Example calculation for our 3-disk run ($N=3$):** $$2^3 - 1 = (2 \times 2 \times 2) - 1 = 8 - 1 = 7 \text{ moves}$$

---

### 🚨 Big-O Performance Conclusion
Because the required steps double with every single added disk, the algorithm scales exponentially.
* **Time Complexity:** $\mathcal{O}(2^N)$ *(Exponential Time)*
* **Space Complexity:** $\mathcal{O}(N)$ *(Linear Space, tracking the memory depth of the recursive call stack)*

### 🎓 Mathematical Derivation (Recurrence Unrolling)

As shown in the lecture guide, the execution trace of the recurrence relation can be formally unrolled as a summation of expanding recursive steps:

$$T(n) = 1 + 2 \cdot T(n-1)$$

Unrolling this sequence down to the base case $T(1) = 1$ reveals a geometric progression:
$$T(n) = 1 + 2 + 2^2 + 2^3 + \dots + 2^{n-1}$$

Using the **Geometric Series Summation** formula provided in the course syllabus:
$$\sum_{k=0}^{n-1} 2^k = \frac{1 - 2^n}{1 - 2} = \frac{1 - 2^n}{-1} = 2^n - 1$$

### 📌 Conclusion
Both the structural code analysis and the formal algebraic unrolling yield the identical mathematical conclusion: the total number of operations required scales at an exponential rate of **$2^N - 1$**, proving the algorithm's complexity rating is **$\mathcal{O}(2^n)$**.


```
def torres_hanoi_optimized(n: int, source: int, target: int) -> int:
    #A peg is simply one of the vertical sticks or towers where you place the disks.
    """
    Optimized implementation of Towers of Hanoi using modern Python conventions.
    Returns the total move count to empirically demonstrate O(2^n) scaling.
    """
    # Base Case
    if n == 1:
        print(f"Move disk from tower {source} to tower {target}")
        return 1

    # Calculate the auxiliary tower dynamically
    auxiliary = 6 - source - target

    # Step 1: Move N-1 disks to the auxiliary tower
    moves = torres_hanoi_optimized(n - 1, source, auxiliary)

    # Step 2: Move the dominant bottom disk to the target
    print(f"Move disk from tower {source} to tower {target}")
    moves += 1

    # Step 3: Move the N-1 disks from auxiliary to target
    moves += torres_hanoi_optimized(n - 1, auxiliary, target)

    return moves

# Execution profiling for 3 disks
print("--- Execution Trace ---")
total_moves = torres_hanoi_optimized(3, 1, 3)
print(f"\nTotal Moves Executed: {total_moves} (Formula Verification: 2^3 - 1 = {2**3 - 1})")
```

##Devolución de cambio por técnica voraz

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

## Viaje por el rio. Programación dinámica

```
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
